```python
import dataset
from matplotlib import pyplot as plt
import numpy as np

m = 100
xs, ys = dataset.get_beans(m)
```

```python
plt.title("Size-Toxicity Function", fontsize=12)
plt.xlabel("Size")
plt.ylabel("Toxicity")

plt.scatter(xs, ys)

w = 0.1
y_pre = w * xs

plt.plot(xs, y_pre)

plt.show()
```


Predicted y:

$y = w x_0$

So error is:

$e_0 = (y_0 - w x_0)^2$

$e_0 = w^2 x_0^2 - 2 w x_0 y_0 + y_0^2$

Derivative of error with respect to w:

$\frac{d e_0}{d w} = 2 w x_0^2 - 2 x_0 y_0$

```python
# Stochastic Gradient Descent
# AKA batch size of 1

for _ in range(100):
    for i in range(m):
        x = xs[i]
        y = ys[i]
        
        y_pre = w * x
        
        # error = (w * x - y)^2 = w^2 * x^2 - 2wxy + y^2
        
        k = 2 * w * x**2 - 2 * x * y
        alpha = 0.1
        w = w - alpha * k
        plt.clf()
        plt.scatter(xs, ys)
        plt.xlim(0, 1)
        plt.ylim(0, 1.2)
        y_pre = w * xs
        plt.plot(xs, y_pre)
        plt.pause(0.01)

plt.show()        
```
