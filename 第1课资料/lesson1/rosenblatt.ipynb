```python
import dataset
from matplotlib import pyplot as plt
```

```python
xs, ys = dataset.get_beans(100)

plt.title("Size-Toxicity Function", fontsize=12)
plt.xlabel("Bean Size")
plt.ylabel("Toxicity")
plt.scatter(xs, ys)
w=0.5

for m in range(100):
    for i in range(100):
        x = xs[i]
        y = ys[i]

        y_pre =  w * x
        error = y-y_pre

        alpha = 0.05
        w = w + alpha * error * x

y_pre =  w * xs

plt.plot(xs, y_pre)
plt.show()
```
